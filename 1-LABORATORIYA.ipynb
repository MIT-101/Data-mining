41-Savol

Quyidagi cellni yurgazing. Kiril-lotin dasturining kodi va natijasini tahlil qiling.

```python
alif={
 'yu':'ю', 'ij':'ы', 'ch':'ч', 'sh':'ш', 'a':'а', 'b':'б', 'c':'с', 'd':'д', 'e':'е',
 'f':'ф', 'g':'г', 'h':'х', 'i':'и', 'j':'ж', 'k':'к', 'l':'л', 'm':'м', 'n':'н',
 'o':'о', 'p':'п', 'q':'к', 'r':'р', 's':'с', 't':'т', 'u':'у', 'v':'в', 'w':'в',
  'x':'х', 'y':'й', 'z':'з'}

alfa = {}
for i,j in alif.items():
  alfa[j]=i


savol = int(input("""Buyruqni tanlang:

      1 - 'lotin-кирил'     0 - 'кирил-lotin' """))


if savol==1:
  l=[]
  soz=input("\n So'zni kiriting: ")
  for i in soz:
    l.append(alif[i])
  print("".join(l))

else:
  k=[]
  soz=input("So'zni kiriting: ")
  for i in soz:
    k.append(alfa[i])
  print("".join(k))
```
